```python
import pickle
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
```

```python
file_names= os.listdir('faces')
```

```python
face_cascade= cv2.CascadeClassifier('cascade.xml')
```

```python
with open('face_projection.pkl', 'rb') as f:
    pca= pickle.load(f)
    
with open('face_validation.pkl', 'rb') as f:
    clf= pickle.load(f)
```

```python
images= []
for file in file_names:
    img= cv2.imread('faces/'+file, 0)
    img= cv2.resize(img, (500,500))
    images.append(img)
```

```python
count= 0
for n in range(1,44):
    faces= face_cascade.detectMultiScale(images[n], scaleFactor= 1.02, minNeighbors= 3)

    for x,y,w,h in faces:
        count+= 1
        img= images[n][y:y+h,x:x+w]
        plt.imshow(img)
        plt.show()
        cv2.imwrite('only_faces/'+str(count)+'.jpg',img)
        img= cv2.resize(img, (24,24))/255
        projection= pca.transform([img.flatten()])
        pred= clf.predict(projection)
        print (pred)
```
